Creating temp column contains all temperature values

Importing required libraries and modules

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
```

Data spliting is done by training(70),validation(20) and testing(10)

```python
from sklearn.model_selection import train_test_split

# Split unique patient IDs into train and test sets
PT_IDs_train_val, PT_IDs_test = train_test_split(merged_df['patient_id'].unique(), test_size=0.1, random_state=42)
PT_IDs_train, PT_IDs_val = train_test_split(PT_IDs_train_val, test_size=0.222, random_state=42)  # 0.222 = 20% / (70% + 20%)

# Filter data based on patient IDs
train_data = merged_df[merged_df['patient_id'].isin(PT_IDs_train)]
val_data = merged_df[merged_df['patient_id'].isin(PT_IDs_val)]
test_data = merged_df[merged_df['patient_id'].isin(PT_IDs_test)]

# Ensure patient IDs are mutually exclusive
assert len(set(train_data['patient_id']).intersection(set(val_data['patient_id']))) == 0
assert len(set(train_data['patient_id']).intersection(set(test_data['patient_id']))) == 0
assert len(set(val_data['patient_id']).intersection(set(test_data['patient_id']))) == 0

# Split into features and target for each dataset
X_train = train_data[features].values
y_train = train_data[target].values.flatten()

X_val = val_data[features].values
y_val = val_data[target].values.flatten()

X_test = test_data[features].values
y_test = test_data[target].values.flatten()
```

```python
print("Training Data Samples:", len(train_data))
print("Validation Data Samples:", len(val_data))
print("Testing Data Samples:", len(test_data))
```

data reshaping by sequence length as 24 hours for lstm model building

```python
import numpy as np
from sklearn.model_selection import train_test_split

# Assuming X_train, X_val, X_test, y_train, y_val, y_test are defined

# Define the sequence length (timesteps)
timesteps = 24

# Function to reshape data for LSTM input
def reshape_for_lstm(X, y):
    X_lstm = []
    y_lstm = []

    for i in range(len(X) - timesteps + 1):
        X_lstm.append(X[i:i + timesteps, :])
        y_lstm.append(y[i + timesteps - 1])

    X_lstm = np.array(X_lstm)
    y_lstm = np.array(y_lstm)

    return X_lstm, y_lstm

# Reshape train data
X_train_lstm, y_train_lstm = reshape_for_lstm(X_train, y_train)

# Reshape validation data
X_val_lstm, y_val_lstm = reshape_for_lstm(X_val, y_val)

# Reshape test data
X_test_lstm, y_test_lstm = reshape_for_lstm(X_test, y_test)

# Shapes after reshaping for LSTM
print("\nShapes after reshaping for LSTM:")
print("Shape of X_train_lstm:", X_train_lstm.shape)  # (n_train_samples - timesteps + 1, timesteps, n_features)
print("Shape of y_train_lstm:", y_train_lstm.shape)  # (n_train_samples - timesteps + 1,)

print("Shape of X_val_lstm:", X_val_lstm.shape)  # (n_val_samples - timesteps + 1, timesteps, n_features)
print("Shape of y_val_lstm:", y_val_lstm.shape)  # (n_val_samples - timesteps + 1,)

print("Shape of X_test_lstm:", X_test_lstm.shape)  # (n_test_samples - timesteps + 1, timesteps, n_features)
print("Shape of y_test_lstm:", y_test_lstm.shape)  # (n_test_samples - timesteps + 1,)
```

LSTM Model building

```python
 # Set seed for reproducibility
seed_value = 42
np.random.seed(seed_value)
tf.random.set_seed(seed_value)

#Define model parameters

lstm_units = 50
dense_units = 32
dropout_rate = 0.2
optimizer = 'adam'


# Build the LSTM model
model = Sequential([
    LSTM(units=lstm_units, input_shape=(24, 94)),  # input_shape: (timesteps, features)
    Dense(units=dense_units, activation = 'relu'),
    Dropout(dropout_rate),
    Dense(units=1, activation = 'sigmoid')
])

# Compile the model
model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])

# Print a summary of the model
model.summary()
```

Model training and evaluation

```python
# Set seed for reproducibility
seed_value = 42
np.random.seed(seed_value)
tf.random.set_seed(seed_value)

# Define early stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Train model with early stopping
history = model.fit(X_train_lstm, y_train_lstm, epochs=50, batch_size=32,
                    validation_data=(X_val_lstm, y_val_lstm), callbacks=[early_stopping])

# Evaluate model
test_loss, test_accuracy = model.evaluate(X_test_lstm, y_test_lstm)
print("Test Loss:", test_loss)
print("Test Accuracy:", test_accuracy)
```

MSE evaluation

```python
y_pred = model.predict(X_test_lstm)

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y_test_lstm, y_pred)
print("Mean Squared Error (MSE):", mse)
```

K-fold cross validation

```python
from sklearn.model_selection import KFold

# Define the number of folds for cross-validation
n_splits = 5
kf = KFold(n_splits=n_splits, shuffle=True, random_state=seed_value)

# Define lists to store evaluation metrics for each fold
test_losses = []
test_accuracies = []

# Perform K-fold cross-validation
for train_index, test_index in kf.split(X_train_lstm):
    X_train_fold, X_val_fold = X_train_lstm[train_index], X_train_lstm[test_index]
    y_train_fold, y_val_fold = y_train[train_index], y_train[test_index]



    # Define early stopping
    early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

    # Train model with early stopping
    history = model.fit(X_train_fold, y_train_fold, epochs=50, batch_size=32, validation_data=(X_val_fold, y_val_fold), callbacks=[early_stopping])

    # Evaluate model
    test_loss, test_accuracy = model.evaluate(X_test_lstm, y_test_lstm)
    test_losses.append(test_loss)
    test_accuracies.append(test_accuracy)

# Calculate average evaluation metrics across all folds
avg_test_loss = np.mean(test_losses)
avg_test_accuracy = np.mean(test_accuracies)

print("Average Test Loss:", avg_test_loss)
print("Average Test Accuracy:", avg_test_accuracy)
```

finding threshold value by optimal threshold method

AUC ROC method

```python
# Set seed for reproducibility
seed_value = 42
np.random.seed(seed_value)
tf.random.set_seed(seed_value)

from sklearn.metrics import roc_curve, auc

# Predict probabilities for test set
y_pred = model.predict(X_val_lstm)

# Compute ROC curve and AUC
fpr, tpr, thresholds_roc = roc_curve(y_val_lstm, y_pred)
roc_auc = auc(fpr, tpr)

# Find optimal threshold based on ROC curve (You can choose different strategies here)
optimal_idx_roc = np.argmax(tpr - fpr)
optimal_threshold_roc = thresholds_roc[optimal_idx_roc]

print("Optimal Threshold by ROC Curve Method:", optimal_threshold_roc)
print("AUC-ROC:", roc_auc)
```

maximum f1 score method

```python
from sklearn.metrics import precision_recall_curve

# Compute precision-recall curve
precision, recall, thresholds_pr = precision_recall_curve(y_val_lstm, y_pred)

# Find optimal threshold based on Precision-Recall curve (You can choose different strategies here)
fscore = (2 * precision * recall) / (precision + recall + 1e-8)
optimal_idx_pr = np.argmax(fscore)
optimal_threshold_pr = thresholds_pr[optimal_idx_pr]

print("Optimal Threshold by Precision-Recall Method:", optimal_threshold_pr)
print("Best F1 Score:", fscore[optimal_idx_pr])
```

```python
# Directly find the threshold that maximizes F1 score
optimal_idx_f1 = np.argmax(fscore)
optimal_threshold_f1 = thresholds_pr[optimal_idx_f1]

print("Optimal Threshold by F1 Score Method:", optimal_threshold_f1)
print("Best F1 Score:", fscore[optimal_idx_f1])
```

Calculating early detection time ( time difference between predicted CRS onset and actual CRS onset)

```python
# Predict CRS probability using the LSTM model
crs_probabilities = model.predict(X_val_lstm)

# Define a threshold probability
threshold = 0.49

# Detect CRS onset
crs_onset_detected = False
crs_onset_time = None
for idx, prob in enumerate(crs_probabilities):
    if prob[0] > threshold:  # Comparing individual probability with threshold
        crs_onset_detected = True
        crs_onset_time = X_test[idx][0]
        break

# Calculate early detection time if CRS onset is detected
if crs_onset_detected:
    actual_crs_onset_time = merged_df[merged_df['CRS on date (0 No, 1 Yes)'] == 1]['Hours'].iloc[0]
    early_detection_time = crs_onset_time - actual_crs_onset_time
    print("Early CRS detection time:", early_detection_time)
else:
    print("CRS onset not detected.")
```
