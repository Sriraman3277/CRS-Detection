Calculating the Average monitoring time of all patients

Median monitoring time-13 days

```python
import pandas as pd

# Assuming your DataFrame is named data_df
# Convert 'Date of Sample' column to datetime objects if needed
data_df['Date of Sample'] = pd.to_datetime(data_df['Date of Sample'])

# Group data by patient_id
grouped_data = data_df.groupby('PT_ID')

# Initialize a list to store monitoring time for each patient
monitoring_times = []

# Iterate over each group (each patient)
for patient_id, group_data in grouped_data:
    # Find minimum start time and maximum end time for each patient
    start_time = group_data['Date of Sample'].min()
    end_time = group_data['Date of Sample'].max()

    # Calculate monitoring time duration
    monitoring_time = end_time - start_time

    # Append patient ID and monitoring time to the list
    monitoring_times.append({'patient_id': patient_id, 'Monitoring_Time': monitoring_time})

# Create a DataFrame from the list of monitoring times
monitoring_time_df = pd.DataFrame(monitoring_times)

# Calculate median monitoring time across all patients
median_monitoring_time = monitoring_time_df['Monitoring_Time'].median()

# Print the DataFrame with monitoring time for each patient
print(monitoring_time_df)

# Print median monitoring time
print("Median Monitoring Time:", median_monitoring_time)
```

data reshaping as sequence length of 24 hours

```python
import numpy as np
from sklearn.model_selection import train_test_split

# Assuming X_train, X_val, X_test, y_train, y_val, y_test are defined

# Define the sequence length (timesteps)
timesteps = 24

# Function to reshape data for LSTM input
def reshape_for_lstm(X, y):
    X_lstm = []
    y_lstm = []

    for i in range(len(X) - timesteps + 1):
        X_lstm.append(X[i:i + timesteps, :])
        y_lstm.append(y[i + timesteps - 1])

    X_lstm = np.array(X_lstm)
    y_lstm = np.array(y_lstm)

    return X_lstm, y_lstm

# Reshape train data
X_train_lstm, y_train_lstm = reshape_for_lstm(X_train, y_train)

# Reshape validation data
X_val_lstm, y_val_lstm = reshape_for_lstm(X_val, y_val)

# Reshape test data
X_test_lstm, y_test_lstm = reshape_for_lstm(X_test, y_test)

# Shapes after reshaping for LSTM
print("\nShapes after reshaping for LSTM:")
print("Shape of X_train_lstm:", X_train_lstm.shape)  # (n_train_samples - timesteps + 1, timesteps, n_features)
print("Shape of y_train_lstm:", y_train_lstm.shape)  # (n_train_samples - timesteps + 1,)

print("Shape of X_val_lstm:", X_val_lstm.shape)  # (n_val_samples - timesteps + 1, timesteps, n_features)
print("Shape of y_val_lstm:", y_val_lstm.shape)  # (n_val_samples - timesteps + 1,)

print("Shape of X_test_lstm:", X_test_lstm.shape)  # (n_test_samples - timesteps + 1, timesteps, n_features)
print("Shape of y_test_lstm:", y_test_lstm.shape)  # (n_test_samples - timesteps + 1,)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

# Set seed for reproducibility
seed_value = 42
np.random.seed(seed_value)
tf.random.set_seed(seed_value)

# Define the number of folds for cross-validation
n_splits = 5
kf = KFold(n_splits=n_splits, shuffle=True, random_state=seed_value)

# Define lists to store evaluation metrics for each fold
test_losses = []
test_accuracies = []

# Define model parameters
lstm_units = 50
dense_units = 32
dropout_rate = 0.2
optimizer = 'adam'

# Build the LSTM model
model = Sequential([
    LSTM(units=lstm_units, input_shape=(24, X_train_lstm.shape[2])),
    Dense(units=dense_units, activation='relu'),
    Dropout(dropout_rate),
    Dense(units=1, activation='sigmoid')
])

# Compile the model
model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])

# Perform K-fold cross-validation
for fold_idx, (train_index, val_index) in enumerate(kf.split(X_train_lstm)):
    print(f"Fold {fold_idx + 1}/{n_splits}")
    X_train_fold, X_val_fold = X_train_lstm[train_index], X_train_lstm[val_index]
    y_train_fold, y_val_fold = y_train[train_index], y_train[val_index]

    # Define early stopping
    early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

    # Train the model with early stopping
    history = model.fit(X_train_fold, y_train_fold, epochs=50, batch_size=32,
                        validation_data=(X_val_fold, y_val_fold), callbacks=[early_stopping], verbose=0)

    # Evaluate the model on the test set
    test_loss, test_accuracy = model.evaluate(X_test_lstm, y_test_lstm)
    test_losses.append(test_loss)
    test_accuracies.append(test_accuracy)

# Calculate average evaluation metrics across all folds
avg_test_loss = np.mean(test_losses)
avg_test_accuracy = np.mean(test_accuracies)

print("\nK-Fold Cross-Validation Results:")
print(f"Average Test Loss: {avg_test_loss}")
print(f"Average Test Accuracy: {avg_test_accuracy}")
```
